# DiGraph

```python
G = nx.DiGraph()
G.add_edge(0,1,label='r')
G.add_edge(0,3,label='y')
G.add_edge(0,2,label='g')
G.add_edge(1,2,label='g')
G.add_edge(4,0,label='r')
color = [G[u][v]['label'] for u,v in G.edges()]
nx.draw_spring(G,edge_color = color,with_labels=True)      # spring方式画图

Gu = G.to_undirected()
# nx.draw_spring(Gu,edge_color = color)
```

```python
count = 0
for u in G.nodes():
    edge_in = set()
    edge_out = set()
    # 作为出边时
    for v in G[u]:
        edge_out.add(G[u][v]['label'])
    # 作为入边时
    for w in Gu[u]:
        if w in G[u]:
            continue
        edge_in.add(G[w][u]['label'])
    if len(edge_in) ==0 or len(edge_out) ==0:
        continue
    if edge_in != edge_out:
        print('dif node:',u)
        count += 1

print('all nodes:',len(G.nodes()))
print('count:',count)
print('diff rate',count/len(G.nodes()))
```
